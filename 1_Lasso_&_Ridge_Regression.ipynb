* Before understanding Lasso & RidgeRegression you must know about the Linear Regression
* As you can see in the above figure Black line show the Linear Regression
* Linear Regression model focus on best fit line
### when our model fit for best fit line then it may be possible that our model get overfit
* If our moder is overfit then we get error in test
* So in this case we have to reduce this error
* For this we have to generalize our Linear regression model and we have to penalize it
* So for penalise the model we use regularization using L1 and L2
* By the help of Lasso and Ridge we genaralize Linear Regression we can get less error in test at the time of Deployement

###  Lasso Regression
1.  First build the linear regression model
2.  Then we have to calculate cost function means we subtract the actual value and the pridicted value and the squre them
3.  our main task is to make cost value is zero so at this time this is the chance that our linear model may be in overfit situation
4.  So at this time in our cost function we **add lambda and take the magnitude of the slope** . 
* In scikit learn lambda is called alpha.
* The value of alpha is 1 by default and we can adjust it.
5.  so that our linear model can be generalize

### Let's do some hands on the Above technique

```
import pandas as pd
import numpy as np
```

```
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Linear Regression/Lasso & Ridge/bangalore house price prediction OHE-data.csv')
df.head()
```

* Split data set into X(independant) and Y (Dependant)

```
X=df.drop('price',axis=1)
Y=df['price']
print('shape of X',X.shape)
print('shape of Y',Y.shape)
```

* Now break data into X_train,X_test,y_train,y_test
* For this purpose use 
  * from sklearn.model_selection import train_test_split

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=51)
print('shape of X_train',X_train.shape)
print('shape of X_test',X_test.shape)
print('shape of y_tarin',y_train.shape)
print('shape of y_test',y_test.shape)
```

now scale the data

```
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

### Bulid Linear Regression model

```
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```
lr.predict([X_test[0,:]])
```

```
lr.predict(X_test)
```

```
lr.score(X_test,y_test)
```

### Implementing Lasso & Ridge Regression

```
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

Now create object for Ridge
* There are many parameter but we are working on lambda means alpha
* By default value of alpha is 1 you can change it accordingly
* But please don't make alpha 0
* Here we are taking all the aparmeter value default

```
rd=Ridge()
rd.fit(X_train,y_train)
rd.score(X_test,y_test)
```

* if you obsrve linear regression gives us 0.7903837092682249
* if you obsrve Ridge regression gives us 0.790568637433663
* There is some improvement in the accuracy
* But not so good so we will go for **Lasso Regression**

## Lasso Regression

```
ls=Lasso()
ls.fit(X_train,y_train)
ls.score(X_test,y_test)
```

* if you obsrve linear regression gives us 0.7903837092682249
* if you obsrve Ridge regression gives us 0.790568637433663
* if you obsrve Lasso regression gives us 0.8036373003525775
* There is good improvement in the accuracy with lasso

### Now change the value of **alpha parameter** in Ridge & Lasso model

```
rd1=Ridge(alpha=2)
rd1.fit(X_train,y_train)
rd1.score(X_test,y_test)
```

* If you observe that there is not so much impact on the acccuracy
* using alpha value 2 in Ridge
* Now change the alpha value in **Lasso**

```
ls1=Lasso(alpha=2)
ls1.fit(X_train,y_train)
ls1.score(X_test,y_test)
```

* Now we will increase the value of alpha(3) in lasso
* Because Lasso gives us good accuracy

```
ls2=Lasso(alpha=3)
ls2.fit(X_train,y_train)
ls2.score(X_test,y_test)
```

```
ls_score=[]
for i in range(1,10):
  ls1=Lasso(alpha=i)
  ls1.fit(X_train,y_train)
  #ls1.score(X_test,y_test)
  ls_score.append(ls1.score(X_test,y_test))
ls_score
```

* At which alpha value it gives good accuracy you can check
* After that it is decreasing
* so we will choose that alpha value in production

### For Ridge tune the hyperameter 

```
rd_score=[]
for i in range(1,10):
  rd2=Lasso(alpha=i)
  rd2.fit(X_train,y_train)
  #ls1.score(X_test,y_test)
  rd_score.append(rd2.score(X_test,y_test))
rd_score
```
